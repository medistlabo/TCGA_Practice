```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "patient": ["P1","P2","P3","P4","P5","P6","P7","P8","P9","P10","P11","P12"],
    "expression": [12, 35, 18, 40, 28, 15, 22, 9, 31, 14, 45, 11],
    "survival": [40, 10, 30, 8, 15, 35, 20, 45, 12, 38, 6, 42]
}

df = pd.DataFrame(data)

df
```

```python
df["group"] = df["expression"].apply(
    lambda x:"High" if x>=20 else "Low"
)

df
```

```python
high = df[df["group"] == "High"]

high
```

```python
df.groupby("group")["survival"].mean()
```

```python
high_survival = df[df["group"] == "High"]["survival"]
low_survival = df[df["group"] == "Low"]["survival"]

plt.boxplot([high_survival, low_survival])

plt.xticks([1, 2], ["High","Low"])
plt.ylabel("survival")

plt.show()
```

```python
plt.hist(df["expression"])
plt.show()
```

```python
plt.scatter(df["expression"], df["survival"])

plt.show()
```
