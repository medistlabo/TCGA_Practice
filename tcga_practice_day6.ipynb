```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "gene": ["TP53", "EGFR", "KRAS", "MYC", "BRAF", "PIK3CA"],
    "expression": [12.5, 30.1, 18.2, 25.4, 40.2, 15.3],
    "survival": [24, 12, 36, 18, 10, 30],
    "group": ["High", "Low", "High", "Low", "Low", "High"]
}

df = pd.DataFrame(data)

df
```

```python
df[df["expression"] > 20]
```

```python
df[(df["expression"] > 15) & (df["survival"] > 20)]
```

```python
df[df["group"] == "Low"]["expression"].mean()
```

```python
plt.hist(df["expression"])

plt.xlabel("expression")
plt.ylabel("count")

plt.show()
```

```python
df.groupby("group")[["expression" , "survival"]].mean()
```

```python
df.sort_values("survival", ascending = False)
```
