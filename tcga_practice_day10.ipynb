```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "patient": ["P1","P2","P3","P4","P5","P6","P7","P8","P9","P10"],
    "expression": [12, 35, 18, 40, 28, 15, 22, 9, 31, 14],
    "survival": [40, 10, 30, 8, 15, 35, 20, 45, 12, 38]
}

df = pd.DataFrame(data)

df
```

```python
df["group"] = df["expression"].apply(
    lambda x: "High" if x>=20 else "Low"
)
```

```python
df_tcga_high = df[df["group"] == "High"]
df_tcga_high
```

```python
import numpy as np
```

```python
df.groupby("group")["survival"].mean()
```

```python
import matplotlib.pyplot as plt
```

```python
High_survival = df[df["group"] == "High"]["survival"]
Low_survival = df[df["group"] == "Low"]["survival"]

plt.boxplot([High_survival, Low_survival])
plt.show()
```
