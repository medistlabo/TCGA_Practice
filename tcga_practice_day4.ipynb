```python
import pandas as pd

data = {
    "gene": ["TP53", "EGFR", "KRAS", "MYC", "BRAF"],
    "expression": [12.5, 30.1, 18.2, 25.4, 40.2],
    "survival_months": [24, 12, 36, 18, 10],
    "group": ["High", "Low", "High", "Low", "Low"]
}

df = pd.DataFrame(data)

df
```

```python
df[["gene","expression"]]
```

```python
df["expression_x2"] = df["expression"] * 2
df
```

```python
df["high_expression"] = df["expression"] > 28
df
```

```python
df.groupby("group")["survival_months"].mean()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["expression"], df["survival_months"])
plt.xlabel("Expression")
plt.ylabel("Survival")
plt.show()
```

```python
df.to_csv("tcga_like_data.csv")
```
