```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "patient": ["P1","P2","P3","P4","P5","P6","P7","P8"],
    "expression": [12, 35, 18, 40, 28, 15, 22, 9],
    "survival": [40, 10, 30, 8, 15, 35, 20, 45]
}

df = pd.DataFrame(data)

df
```

```python
df["group"] = df["expression"] >= 20
df
```

```python
df["group"] = df["expression"].apply(
    lambda x: "High" if x >= 20 else "Low"
)
```

```python
df.groupby("group")["survival"].mean()
```

```python
import matplotlib.pyplot as plt
```

```python
High_survival = df[df["group"] == "High"]["survival"]
Low_survival = df[df["group"] == "Low"]["survival"]

plt.boxplot([High_survival, Low_survival])

plt.xticks([1 ,2], ["High", "Low"])

plt.xlabel("group")
plt.ylabel("survival")

plt.show()
```
