```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "gene": ["TP53", "EGFR", "KRAS", "MYC", "BRAF"],
    "expression": [12.5, 30.1, 18.2, 25.4, None],
    "survival": [24, 12, 36, 18, 10],
    "group": ["High", "Low", "High", "Low", "Low"]
}

df = pd.DataFrame(data)

df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
clean_df  = df.dropna()

clean_df
```

```python
clean_df.groupby("group")["expression"].mean()
```

```python
clean_df.boxplot(column="expression", by = "group")
plt.show()
```

```python
clean_df["expression"].corr(clean_df["survival"])
```

```python
plt.scatter(clean_df["expression"], clean_df["survival"])

plt.xlabel("expression")
plt.ylabel("suevival")

plt.show()
```

```python
clean_df.to_csv("clean_tcga_like_data.csv")
```
