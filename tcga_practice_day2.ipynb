```python
import pandas as pd
from lifelines import KaplanMeierFitter
import matplotlib.pyplot as plt
```

```python
data = {
    "time": [5, 6, 6, 2, 4, 3],
    "event": [1, 0, 1, 1, 0, 1],
    "group": ["High", "High", "Low", "Low", "High", "Low"]
}

df = pd.DataFrame(data)
```

```python
df.groupby("group")["time"].mean()
```

```python
high = df[df["group"] == "High"]
high



```

```python
high = df[df["group"] == "High"]

kmf = KaplanMeierFitter()

kmf.fit(high["time"], high["event"], label = "High")
kmf.plot()

plt.show()
```
